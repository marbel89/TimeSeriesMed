We want to find seasonal patterns and the order of AR (Autoregressive) and MA (Moving Average). We can do that by plotting the (partial) autocorrelation function.

ACF:

Decay Pattern:

    MA(q) Process: For a pure MA(q) process, the ACF will show significant spikes at lag 1 through q, and then it will drop to zero.
    AR(p) Process: For a pure AR(p) process, the ACF will show a gradual decay, either exponentially or in a damped sine wave pattern.
    ARMA(p,q) Process: For an ARMA(p,q) process, the ACF will exhibit a mixture of these patterns.

In general, we want to look for significant spikes outside the confidence intervals.

Next step is to assess if the data is stationary via the augmented Dickey-Fuller (ADF) test.

```python
adfuller_result = adfuller(df.y)
print(f"ADF statistic: {adfuller_result[0]}")
print(f"p-value: {adfuller_result[1]}")
```

Rejecting the null hypothesis (i.e. the timeseries is **non-stationary**) with a p-value of 1.
Stationarity: Mean, variance, autocorrelation... do not change overtime.
This means we apply first order differencing and check again.

```python
y_diff = np.diff(df.y, n=1)
adfuller_result = adfuller(y_diff)
print(f"ADF statistic: {adfuller_result[0]}")
print(f"p-value: {adfuller_result[1]}")
```

Since the p-value is still over the statistical convention 0.05, the time series still isn't stationary.
Non-stationary time series data typically have two main components that need to be addressed: Trend and seasonality.

First-order differencing is used to remove the trend by subtracting the current observation from the previous observation.

Seasonal differencing is used to remove seasonal effects by subtracting the current observation from the observation at the same point in the previous season (e.g., 12 months ago for monthly data). Why Not Directly Apply Seasonal Differencing? First-order differencing helps to remove the linear trend component, making the seasonal differencing more effective. Trend and seasonality can interact in complex ways. Removing the trend first simplifies the data structure, making it easier to isolate and remove the seasonal component.


```python
y_diff_season = np.diff(y_diff, n=12)
adfuller_result = adfuller(y_diff_season)
print(f"ADF statistic: {adfuller_result[0]}")
print(f"p-value: {adfuller_result[1]}")
```

Our final model will be SARIMA(p,1,q)(P,1,Q)12

d=1: Because you performed first-order differencing.<br>
D=1: Because you performed seasonal differencing.<br>
s=12: Because the seasonal period is 12 months.

Now we need to find optimal values for p, q, P and Q. As a selection criterion, wie use Akaike information criterion. This criterion helps compare models, and if respective models are neither over nor underfitting.

First, we need to split the data into test and train data (we are dealing with 36 months of test data). Then we use the function optimize_SARIMAX from statsmodels to find the best values.

```python
train = df.y[:168]
test = df.y[168:]
 
print(len(test))
```

```python
def optimize_SARIMAX(endog: Union[pd.Series, list], 
                     exog: Union[pd.Series, list], 
                     order_list: list, d: int, D: int, s: int) -> pd.DataFrame:

    results = []
    
    for order in tqdm_notebook(order_list):
        try:
            model = SARIMAX(endog, exog,
                           order=(order[0], d, order[1]),
                           seasonal_order=(order[2], D, order[3], s),
                           simple_differencing=False).fit(disp=False)
        except:
            continue
        
        aic = model.aic
        results.append([order, model.aic])
    
    result_df=pd.DataFrame(results)
    result_df.columns=["(p,q,P,Q)", "AIC"]
    
    result_df = result_df.sort_values(by="AIC", ascending=True).reset_index(drop=True)
    
    model.plot_diagnostics(figsize=(15,12))
    
    return result_df

ps = range(0, 5, 1)
qs = range(0, 5, 1)
Ps = range(0, 5, 1)
Qs = range(0, 5, 1)
 
order_list = list(product(ps, qs, Ps, Qs))
 
d = 1
D = 1
s = 12
    
```

```python
SARIMA_result_df = optimize_SARIMAX(train, None, order_list, d, D, s)
SARIMA_result_df
```

Next step: Ljung-Box test to check if residuals are uncorrelated.


```python
sarima_model = SARIMAX(train, order=(3, 1, 1),
                       seasonal_order=(1, 1, 3, 12),
                       simple_differencing=False)
sarima_model_fit = sarima_model.fit(disp=False)
sarima_model_fit.plot_diagnostics(figsize=(10,8))
                      
        
```

```python
residuals = sarima_model_fit.resid
values = acorr_ljungbox(residuals, np.arange(1, 11,1))
values
```

All p-values are above 0.05, so we do not reject the null hypothesis that residuals are correlated. We can use the model for forecasting.

```python
def rolling_forecast(df: pd.DataFrame, train_len: int, horizon: int, window: int, method: str) -> list:
    total_len = train_len + horizon
    end_idx = train_len
    
    if method == "last_season":
        pred_last_season=[]
        
        for i in range(train_len, total_len, window):
            last_season=df["y"][i-window:i].values
            pred_last_season.extend(last_season)
        return pred_last_season
    
    elif method == "SARIMA":
        pred_SARIMA = []
        
        for i in range(train_len, total_len, window):
            model = SARIMAX(df["y"][:i], order=(p,d,q),seasonal_order=(P,D,Q,s),simple_differencing=False)
            res = model.fit(disp=False)
            predictions = res.get_prediction(0, i + window - 1)
            oos_pred = predictions.predicted_mean.iloc[-window:]
            pred-SARIMA.extend(oos_pred)
        return pred_SARIMA
                
```
